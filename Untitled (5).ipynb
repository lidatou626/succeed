```python
import pandas as pd
import jieba
from snownlp import SnowNLP

# 读取Excel文件中的景区评论数据
df = pd.read_excel('景区与评论数据集.xlsx')

# 定义提取关键词的函数
def extract_keywords(text):
    # 处理缺失值
    if isinstance(text, float) and pd.isna(text):
        return []
    # 使用jieba进行分词
    words = jieba.lcut(text)
    return words

# 定义情感分析函数
def get_sentiment_score(text):
    # 处理缺失值
    if isinstance(text, float) and pd.isna(text):
        return None
    # 使用SnowNLP进行情感分析
    s = SnowNLP(text)
    # 返回情感得分(0-1之间，越接近1表示情感越积极)
    return s.sentiments

# 对每条评论提取关键词并添加到新列
df['关键词'] = df['评论内容'].apply(extract_keywords)

# 对每条评论进行情感分析并添加到新列
df['情感得分'] = df['评论内容'].apply(get_sentiment_score)

# 将处理后的数据保存到新的Excel文件
df.to_excel('景区与评论数据集.xlsx_关键词与情感得分.xlsx', index=False)
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

scenic_reviews = df.groupby('景区名称')['评论内容'].agg(lambda x: ' '.join(x)).reset_index()

vectorizer = TfidfVectorizer(stop_words=None)  # 可添加中文停用词
tfidf_matrix = vectorizer.fit_transform(scenic_reviews['评论内容'])

similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)

scenic_index = {name: i for i, name in enumerate(scenic_reviews['景区名称'])}

def get_similar_scenics(scenic_name, top_n=5):
    if scenic_name not in scenic_index:
        return f"抱歉，景区 '{scenic_name}' 不在数据集中。"
    
    idx = scenic_index[scenic_name]
    
    sim_scores = list(enumerate(similarity_matrix[idx]))
    
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    top_scenics = sim_scores[1:top_n+1]
    
    result = []
    for i, score in top_scenics:
        result.append({
            '景区名称': scenic_reviews.iloc[i]['景区名称'],
            '相似度得分': score
        })
    
    return pd.DataFrame(result)
```

```python
import os
import numpy as np
from PIL import Image
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input
from sklearn.metrics.pairwise import cosine_similarity

class ImageFeatureExtractor:
    def __init__(self, model_name='resnet50'):
        if model_name == 'resnet50':
            self.model = ResNet50(weights='imagenet', 
                                include_top=False, 
                                pooling='avg')
        self.input_size = (224, 224)
    
    def load_image(self, img_path):
        img = Image.open(img_path)
        if img.mode != 'RGB':
            img = img.convert('RGB')
        return img.resize(self.input_size)

    def extract_features(self, img_path):
        try:
            img = self.load_image(img_path)
            x = image.img_to_array(img)
            x = np.expand_dims(x, axis=0)
            x = preprocess_input(x)
            features = self.model.predict(x)
            return features.flatten()
        except Exception as e:
            print(f"Error processing {img_path}: {str(e)}")
            return None

    def process_dataset(self, dataset_dir):
        features = {}
        for root, _, files in os.walk(dataset_dir):
            for file in files:
                if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                    img_path = os.path.join(root, file)
                    feat = self.extract_features(img_path)
                    if feat is not None:
                        features[file.split('.')[0]] = feat
        return features

    def compute_similarity(self, features_dict):

        names = list(features_dict.keys())
        features = np.array([features_dict[name] for name in names])
        similarity_matrix = cosine_similarity(features)
        
        print("景区相似度结果:")
        for i, name in enumerate(names):
            print(f"\n与 {name} 最相似的景区：")
            sim_scores = list(enumerate(similarity_matrix[i]))
            sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:6]
            for idx, score in sim_scores:
                print(f"{names[idx]}: {score:.4f}")
                
        return names, similarity_matrix
```

```python
if __name__ == "__main__":
    try:
        extractor = ImageFeatureExtractor()
        print("✅ 特征提取器初始化完成")
        
        print("⏳ 正在处理图片数据集...")
        features = extractor.process_dataset('images')
        print(f"✅ 成功处理 {len(features)} 个景区图片")
        
        # 添加检查：确保有足够的图片进行相似度计算
        if len(features) < 2:
            raise ValueError(f"需要至少2张图片来计算相似度，当前只有{len(features)}张")
            
        scenic_names, similarity_matrix = extractor.compute_similarity(features)
        np.save('scenic_similarity.npy', similarity_matrix)

        print("\n🏆 整体最相似景区TOP10：")
        pairs = []
        for i in range(len(scenic_names)):
            for j in range(i+1, len(scenic_names)):
                pairs.append((scenic_names[i], scenic_names[j], similarity_matrix[i][j]))
        
        top_pairs = sorted(pairs, key=lambda x: -x[2])[:10]
        for rank, (name1, name2, score) in enumerate(top_pairs, 1):
            print(f"{rank}. {name1} ↔ {name2}  相似度: {score:.2f}")
            
    except ValueError as e:
        print(f"❌ 数据错误: {str(e)}")
    except Exception as e:
        print(f"❌ 程序运行出错: {str(e)}")
```
